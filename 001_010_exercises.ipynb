### <a name='1'></a> Ćwiczenie 1
Sprawdź czy wszystkie elementy tablic $A, B, C$ oraz $D$ zwracają wartość logiczną _True_.



```
A = np.array([[3, 2, 1, 4],
              [5, 2, 1, 6]])

B = np.array([[3, 2, 1, 4],
              [5, 2, 0, 6]])

C = np.array([[True, False, False],
              [True, True, True]])

D = np.array([0.1, 0.3])
```



__Wskazówka:__ Użyj funkcji _np.all()._

````
A = np.array([3, 2, 1, 4, 5, 2, 1, 6])
A = A.reshape(2, -1)
print(f'A =\n{A}\n')

B = np.array([3, 2, 1, 4, 5, 2, 0, 6]).reshape(2, -1)
print(f'B =\n{B}\n')

C = np.array([[True, False, False],
              [True, True, True]])
print(f'C =\n{C}\n')

D = np.array([0.1, 0.3])
print(f'D =\n{D}\n')
````

````
print(f'A = {np.all(A)}\nB = {np.all(B)}\nC = {np.all(C)}\nD = {np.all(D)}')
# np.all() zwraca false gdy w tablicy są wartości false, czyli liczba 0. W przeciwnym razie true.

# Odpowiedź Nie wszystkie elementy tablic A, B, C i D zwracają wartość true.
````
